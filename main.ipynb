```python
import requests
import json

def translate_text(x):
    try:
        translation_result = translate(x, dest='mai', src='en')
        return translation_result
    except requests.exceptions.RequestException as e:
        print(f"Error making translation request: {e}")
        return None  # Handle the error gracefully, you can modify this based on your needs
    except json.decoder.JSONDecodeError as e:
        print(f"Error decoding JSON response: {e}")
        return None  # Handle the error gracefully, you can modify this based on your needs

# Assuming 'max_length_of_news' is defined
df['translated'] = df.tail(max_length_of_news)['full_article'].apply(translate_text)
```

```python
# import transformers for bertclassifier
import torch
import transformers
from transformers import BertTokenizer, AutoTokenizer, BertModel, BertConfig, AutoModel

import torch.nn as nn
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
from huggingface_hub import hf_hub_download

hf_hub_download(
    repo_id="rockerritesh/maithili_classification_model", filename="maibert.bin"
)
```

```python
# Creating the customized model, by adding a drop out and a dense layer on top of distil bert to get the final output for the model.


class BERTClass(torch.nn.Module):
    def __init__(self):
        super(BERTClass, self).__init__()
        self.roberta = AutoModel.from_pretrained('google/muril-base-cased')
        self.l2 = torch.nn.Dropout(0.5)
        self.fc = torch.nn.Linear(768,10)

    def forward(self, ids, mask, token_type_ids):
        _, features = self.roberta(ids, attention_mask = mask, token_type_ids = token_type_ids, return_dict=False)
        output_2 = self.l2(features)
        output = self.fc(output_2)
        return output


# Load the saved state dictionary
model_state_dict_path = '/home/vscode/.cache/huggingface/hub/models--rockerritesh--maithili_classification_model/snapshots/dccb02b7f0025ff44b04191fbe94b0d5006b5885/maibert.bin'
model_state_dict = torch.load(model_state_dict_path,map_location=torch.device('cpu'))

# Create an instance of your model
model = BERTClass()
```

```python
# Load the state dictionary into the model
model.load_state_dict(model_state_dict)
```

```python
target_cols = ['Politics', 'Culture', 'Sports', 'Literature', 'Entertainment',
       'Health', 'EduTech', 'Opinion', 'Interview', 'Economy']
```

```python
tokenizer = AutoTokenizer.from_pretrained('rockerritesh/maiBERT_TF')
```

```python
# Get input from the user
user_input = input("Enter a text for classification: ")

# Tokenize and preprocess the input
tokenized_input = tokenizer.encode_plus(
                  user_input,
                  truncation=True,
                  add_special_tokens=True,
                  max_length=256,
                  padding='max_length',
                  return_token_type_ids=True
              )

# Move tensors to the available device (GPU or CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"
input_ids = torch.tensor(tokenized_input['input_ids']).unsqueeze(0).to(device)  # Add unsqueeze(0) to add batch dimension
attention_mask = torch.tensor(tokenized_input['attention_mask']).unsqueeze(0).to(device)
token_type_ids = torch.tensor(tokenized_input['token_type_ids']).unsqueeze(0).to(device)

model.eval()
# Perform inference
with torch.no_grad():
    output = model(input_ids, attention_mask, token_type_ids)
print('output_classes',target_cols[torch.argmax(output)])
```
